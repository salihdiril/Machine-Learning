```python
import cv2
import os 
import numpy as np
import matplotlib.pyplot as plt
import math
from PIL import Image
import random
import matplotlib.image as mpimg
from imutils import paths
import sys
import operator
```

```python
def get_paths_of_lines_and_pages_v4(book_path):
  book = sorted(os.listdir(book_path), key = lambda x: int(x))
  sheet_paths = []
  for page in book:
    sheet_paths.append(book_path+"/"+page)

  line_paths = []
  for page in sheet_paths:
    page_lines = sorted(os.listdir(page), key = lambda x: int(x.split(".")[0]))
    line_path = []
    for line in page_lines:
      line_path.append(page+"/"+line)
    line_paths.append(line_path)

  return {
      "sheet_paths": sheet_paths,
      "line_paths": line_paths
  }
```

```python
pages_path = "/content/drive/MyDrive/Line Extraction/Images/Image Lines/evliya_celebi_2v2"
book_paths = get_paths_of_lines_and_pages_v4(pages_path)
```

```python
def fill_page_nums_with_zero(book_paths):
  for page in book_paths["line_paths"]:
      for line in page:
        line_num = line.split("/")[-1].split(".")[0]
        parent_path = "/".join(line.split("/")[:-1])
        if len(line_num) == 4:
          new_num = "000" + line_num
          path = parent_path + "/" + new_num + ".png"
          os.rename(line, path)
        if len(line_num) == 5:
          new_num = "00" + line_num
          path = parent_path + "/" + new_num + ".png"
          os.rename(line, path)
        if len(line_num) == 6:
          new_num = "0" + line_num
          path = parent_path + "/" + new_num + ".png"
          os.rename(line, path)
```

```python
def replace_zeros_with_hyphen(book_paths):
  for page in book_paths["line_paths"]:
      for line in page:
        line_num = line.split("/")[-1].split(".")[0]
        parent_path = "/".join(line.split("/")[:-1])
        line_num_list = list(line_num)
        line_num_list[4]= "-"
        line_num_with_hyphen = "".join(line_num_list)
        new_line = parent_path + "/" + line_num_with_hyphen + ".png"
        os.rename(line, new_line)
```
