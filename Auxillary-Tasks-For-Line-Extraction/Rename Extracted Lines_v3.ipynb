```python
def get_paths_of_lines_and_pages(book_path):
  book = sorted(os.listdir(book_path), key = lambda x: int(x.split(" ")[0].split("-")[1]))
  sheet_paths = []
  subpages = []
  lines = []
  for page in book:
    sheet_paths.append(book_path+"/"+page)

  for path in sheet_paths:
    page_parts = sorted(os.listdir(path), reverse=True)
    page_parts_paths = []
    for part in page_parts:
      page_parts_paths.append(path+"/"+part)
    subpages.append(page_parts_paths)

  for page in subpages:
    sheet = []
    for part in page:
      page_lines = sorted(os.listdir(part), key = lambda x: int(x.split("/")[-1].split("_")[-1].split(".")[0]))
      lines_path = []
      for line in page_lines:
        lines_path.append(part+"/"+line)
      sheet.append(lines_path)
    lines.append(sheet)

  return {
      "sheet_paths": sheet_paths,
      "subpage_paths": subpages,
      "line_paths": lines
  }
```

```python
def rename_paths(book_path, parent_folder, path_dict, starting_page):
  if len(parent_folder.split("/")) - len(book_path.split("/")) == 0:
    i = starting_page
    for page in path_dict["sheet_paths"]:
      new_filename = book_path + "/" + str(i)
      os.rename(page, new_filename)
      i += 1

  elif len(parent_folder.split("/")) - len(book_path.split("/")) == 1:
    for page in path_dict["subpage_paths"]:
      i = int(page[0].split("/")[-2])*100 + 1
      for part in page:
        new_filename = book_path + "/" + part.split("/")[-2] + "/" + str(i)
        os.rename(part, new_filename)
        i += 1

  elif len(parent_folder.split("/")) - len(book_path.split("/")) == 2:
    for page in path_dict["line_paths"]:
      for part in page:
        i = int(part[0].split("/")[-2])*1000 + 1
        for line in part:
          if line.split("/")[-2] == int(line.split("/")[-3])*100+1  and (i >= int(line.split("/")[-3])*100000 and i < int(line.split("/")[-3])*100000 + 1999):
            i += 1000
          elif line.split("/")[-2] == int(line.split("/")[-3])*100+2 and (i >= int(line.split("/")[-3])*100000 and i < int(line.split("/")[-3])*100000 + 2999):
            i += 2000
          new_filename = book_path + "/" + line.split("/")[-3] + "/" + line.split("/")[-2] + "/" + str(i)
          os.rename(line, new_filename)
          i += 1

  else:
    print("Parent folder isn't a page, subpage or line folder!!!")
```
