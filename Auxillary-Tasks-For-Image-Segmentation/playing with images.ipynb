```python
img = cv2.imread("/content/drive/MyDrive/Datasets/mask-image/original/1.png")
img.shape
```

```python
height, width, depth = img.shape
window_size = 480
resized_img_height = (height // window_size)*window_size
resized_img_width = (width // window_size)*window_size
resized_img = cv2.resize(img, (resized_img_width, resized_img_height))
resized_img.shape
```

```python
height2, width2, depth2 = resized_img.shape

row_num = height2 // window_size
col_num = width2 // window_size
```

```python
row_num
```

```python
col_num
```

```python
patches = []
for i in range(row_num):
        for j in range(col_num):
            patch = resized_img[i*window_size: (i+1)*window_size, j*window_size: (j+1)*window_size, :]
            patches.append(patch)
```
