```python
def concat(arr, n):
  patch = np.concatenate(([arr[i] for i in range(n)]), axis=1)
  height, width, depth = patch.shape
  figsize = width / float(dpi), height / float(dpi)

  return patch, figsize

def concat_v2(arr):
  patch = np.concatenate(([arr[i] for i in range(len(arr))]), axis=1)
  height, width, depth = patch.shape
  figsize = width / float(dpi), height / float(dpi)

  return patch, figsize

def concat_axis0(arr, n):
  patch = np.concatenate(([arr[i] for i in range(n)]), axis=0)
  height, width, depth = patch.shape
  figsize = width / float(dpi), height / float(dpi)

  return patch, figsize

def plot(img, figsize):
  plt.figure(figsize=figsize)
  plt.imshow(img)
  plt.show()
```

```python
def create_patches(path ,window_size):
  patches = []

  img = cv2.imread(path)
  height, width, depth = img.shape

  row_num = height // window_size
  col_num = width // window_size

  for i in range(row_num+1):
    for j in range(col_num+1):
      patch = img[i*window_size: (i+1)*window_size, j*window_size: (j+1)*window_size, :]
      patches.append(patch)

  return patches
```

```python
def img_show(path):
  dpi = matplotlib.rcParams['figure.dpi']
  img = cv2.imread(path)
  height, width, depth = img.shape
  figsize = width / float(dpi), height / float(dpi)

  plt.figure(figsize=figsize)
  plt.imshow(img)
  plt.show()
```

```python
def resize_to_window_scale(img, window_size):
  height, width, depth = img.shape
  resized_img_height = (height // window)*window
  resized_img_width = (width // window)*window
  resized_img = cv2.resize(img, (resized_img_width, resized_img_height))
  return resized_img
```

```python
def create_patches_from_img(img ,window_size):
  patches = []

  height, width, depth = img.shape

  row_num = height // window_size
  col_num = width // window_size

  for i in range(row_num):
    for j in range(col_num):
      patch = img[i*window_size: (i+1)*window_size, j*window_size: (j+1)*window_size, :]
      patches.append(patch)

  return patches
```

```python
def concat(arr):
  patch = np.concatenate(([arr[i] for i in range(len(arr))]), axis=1)
  return patch

def concat_axis0(arr):
  patch = np.concatenate(([arr[i] for i in range(len(arr))]), axis=0)
  return patch

def plot(img, figsize):
  plt.figure(figsize=figsize)
  plt.imshow(img)
  plt.show()

def concat_page_rows(patches, img, window_size):
  rows = []
  height, width, _ = img.shape
  row = height // window_size
  col = width // window_size
  idx = 0
  for i in range(row):
    rows.append(concat(patches[idx:idx+col]))
    idx += col

  return rows

def concat_page(rows):
  page = concat_axis0([rows[i] for i in range(len(rows))])
  return page


def img_show_from_arr(img):
  dpi = matplotlib.rcParams['figure.dpi']
  height, width, depth = img.shape
  figsize = width / float(dpi), height / float(dpi)

  plt.figure(figsize=figsize)
  plt.imshow(img)
  plt.show()

def from_patches_to_image(patches, img, window_size):
  rows = concat_page_rows(patches, img, window_size)
  page = concat_page(rows)
  return page
```
