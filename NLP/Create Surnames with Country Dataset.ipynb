```python
target_file = "/content/drive/MyDrive/Datasets/RNN/data/names.txt"
```

```python
import glob
files = glob.glob("/content/drive/MyDrive/Datasets/RNN/data/names/*.txt")
```

```python
for file in files:
  # open both files
  with open(file, "r") as source, open(target_file, "a") as target:

    # read content from source file
    for line in source:

      # append content to target file
      target.write(line)
```

```python
with open(target_file, "r") as fp:
  lines = len(fp.readlines())
  print('Total Number of lines:', lines)
```

```python
import os

countries = "/content/drive/MyDrive/Datasets/RNN/data/countries.txt"

for file in files:
  with open(file, "r") as fp1, open(countries, "a") as fp2:
    num_lines = len(fp1.readlines())
    country_name = os.path.splitext(os.path.basename(file))[0] + "\n"
    for i in range(num_lines):
      fp2.write(country_name)
```
